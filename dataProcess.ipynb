## 4 - Opérations sur le fichier CSV burkina_location.csv

### 4.1 - extrayons les données contenant le nom 'gounghin', enregistrons-le sous le fichier gounghin.csv

```python
burkina_location = pd.read_csv("burkina_location.csv", index_col=False)
burkina_location
```

```python
hello = [burkina_location.values[v] for v in burkina_location.index.values if "Gounghin".lower() in burkina_location.values[v][1].lower()]
#hello
#gounghin = burkina_location.loc[burkina_location["location_name"]=="Gounghin"]
gounghin = pd.DataFrame(hello, columns=burkina_location.columns)
gounghin
```

```python
gounghin.to_csv("gounghin.csv", index=False)
```

### 4.2 - extrayons la sous-partie de la base de données (fichier burkina_location.csv), dont les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)

```python
a_to_P = burkina_location.loc[burkina_location["location_name"] <= "P"]
a_to_P_sorted = a_to_P.sort_values(by="location_name").reset_index()
a_to_P_sorted.drop(columns="index", inplace=True)
a_to_P_sorted.drop(index=0, inplace=True)
a_to_P_sorted
```

### 4.3 - Identifions respectivement, la latitude minimale, la longitude minimale et les noms de lieux correspondants

```python
lat_min = burkina_location['lat'].min()
lon_min = burkina_location['long'].min()
#print("lat_min = ", lat_min, "\nlon_min = ", lon_min)

print("Lieu correspondant à la latitude minimale : \n", burkina_location.loc[burkina_location['lat']==lat_min],
    "\n\nLieu correspondant à la longitude minimale : \n", burkina_location.loc[burkina_location['long']==lon_min]
     )
```

### 4.3 - Lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)

```python
lieux_lat_sup_11 = burkina_location.loc[burkina_location['lat']>=11]
lieux_lat_sup_11_et_long_sup_0p5 = lieux_lat_sup_11.loc[lieux_lat_sup_11['long']<=0.5]
lieux_lat_sup_11_et_long_sup_0p5
```

```python
lieux_lat_sup_11_et_long_sup_0p5['lat'].min()
```

```python
lieux_lat_sup_11_et_long_sup_0p5['long'].max()
```
